# Younger age rotating models are degenerate w/ older non-rotating models
## (sort of...)
### $\frac{\Omega}{\Omega_c} \uparrow \approx \tau \uparrow$

In MIST, as mentioned below, models are still quite a bit redder along the MSTO than non-rotating models, due to gravity darkening. However, core luminosity enhancement is able to make the models generally brighter than non-rotating models still at later stages of MSTO. These isochrones may be degenerate with older non-rotating models.

Below is showing that you could confuse an ~80 Myr isochrone that is rotating with a non-rotating 70 Myr isochrone, meaning that essentially, one might say age $\tau$ has increased due to including rotation.

```python
rots = [0.0, 0.0, 0.6]
num_isos=len(rots)
fehs = [0.15]*num_isos
ages = [7.83,7.9,7.9]
datapath = ['/n/home12/sgossage/match2.6_updated/MIST/Hyades/photometry/hyades_debruTIGS_TychoBV.phot']#,
            #'/n/home12/sgossage/match2.6_updated/MIST/Hyades/photometry/hyades_kopytova16_2MASSJK.phot']
filters = [('Tycho_B','Tycho_V')]#, ('2MASS_J', '2MASS_Ks')]
pts, afig, anax = pgc.compare_iso(fehs, rots, ages, filters, dataf=[None], dmod=[3.34], ebv = [0.001])
img = mpimg.imread(os.path.join(os.getcwd(), 'plots', 'IsoCMD_comparison.png'))
plt.figure(figsize=(24,14))
plt.imshow(img)
plt.axis('off')
```

# Older age rotating models are degenerate w/ younger non-rotating models
### $\frac{\Omega}{\Omega_c} \uparrow = \tau \downarrow$

At older ages, ~100 Myr+, gravity darkening dominates core luminosity enhancement due to rotation. Stars appear significantly more red than non-rotating models, and less luminous as they move along MSTO. Gravity darkening makes stars more red via von Zeipel stuff -- lower effective gravity at equator, more oblate, more expanded & equatorial cooler material than at the pole.

```python
rots = [0.0, 0.0, 0.6]
num_isos=len(rots)
fehs = [0.15]*num_isos
ages = [8.73,8.8,8.8]
datapath = ['/n/home12/sgossage/match2.6_updated/MIST/Hyades/photometry/hyades_debruTIGS_TychoBV.phot']#,
            #'/n/home12/sgossage/match2.6_updated/MIST/Hyades/photometry/hyades_kopytova16_2MASSJK.phot']
filters = [('Tycho_B','Tycho_V')]#, ('2MASS_J', '2MASS_Ks')]
pts, afig, anax = pgc.compare_iso(fehs, rots, ages, filters, dataf=[None], dmod=[3.34], ebv = [0.001])
img = mpimg.imread(os.path.join(os.getcwd(), 'plots', 'IsoCMD_comparison.png'))
plt.figure(figsize=(24,14))
plt.imshow(img)
plt.axis('off')
```

### At older ages, younger rotating models may be confused w/ older non-rotating models...
So at older ages, due to the much stronger reddening (significantly weaker rotational mixing, less luminosity), increasing rotation will confuse a *younger* rotating model with an *older* non-rotating model. Thus, one might say that including rotation in models at this age will *decrease* age $\tau$; i.e. rotation has lead to a best-fit age that is younger than the best-fit non-rotating age. (This is shown below.) 

```python
rots = [0.0, 0.0, 0.4]
num_isos=len(rots)
fehs = [0.15, 0.15, 0.05]
ages = [8.73,8.69,8.69]
datapath = ['/n/home12/sgossage/match2.6_updated/MIST/Hyades/photometry/hyades_debruTIGS_TychoBV.phot']#,
            #'/n/home12/sgossage/match2.6_updated/MIST/Hyades/photometry/hyades_kopytova16_2MASSJK.phot']
filters = [('Tycho_B','Tycho_V')]#, ('2MASS_J', '2MASS_Ks')]
pgc.compare_iso(fehs, rots, ages, filters, dataf=[None], dmod=[3.34], ebv = [0.001])
img = mpimg.imread(os.path.join(os.getcwd(), 'plots', 'IsoCMD_comparison.png'))
plt.figure(figsize=(24,14))
plt.imshow(img)
plt.axis('off')
```

However, decreasing metallicity can also cause a degeneracy where a lower metallicity but more strongly rotation model can mimic a higher metallicity non-rotating model.

```python
rots = [0.0, 0.0, 0.6]
num_isos=len(rots)
fehs = [0.15, 0.15, 0.00]
ages = [8.73,8.8,8.73]
datapath = ['/n/home12/sgossage/match2.6_updated/MIST/Hyades/photometry/hyades_debruTIGS_TychoBV.phot']#,
            #'/n/home12/sgossage/match2.6_updated/MIST/Hyades/photometry/hyades_kopytova16_2MASSJK.phot']
filters = [('Tycho_B','Tycho_V')]#, ('2MASS_J', '2MASS_Ks')]
pgc.compare_iso(fehs, rots, ages, filters, dataf=[None], dmod=[3.34], ebv = [0.001])
img = mpimg.imread(os.path.join(os.getcwd(), 'plots', 'IsoCMD_comparison.png'))
plt.figure(figsize=(24,14))
plt.imshow(img)
plt.axis('off')
```

So at older ages, the picture is a bit more complicated...

+ Rotating models become more red the quicker that they rotate.
+ At these ages, rotational mixing is relatively weak, and so the core luminosity is enhanced less.
+ Lowering [Fe/H] will shift the model blueward, reducing this effect somewhat.
+ Decreasing age will increase luminosity and shift the MSTO blueward.

On the other hand, increasin rotation makes the models more red a the MSTO, so other than metallicity, we might change the age of the cluster. The MS maintains a good fit, and while the increased luminosity may decrease the fit quality overall, it may be a more favorable decrease than that acquired when failling to fit the MS very well (as with decreased luminosity.)

```python
rots = [0.0, 0.6]
num_isos=len(rots)
fehs = [0.0, 0.0]
ages = [8.2, 8.2]
pgc.compare_iso(fehs, rots, ages, [('log_Teff','log_L')], pltmass=False, geneva_on = True, ylim= [0.95, 3.0], xlim = [4.2, 3.85], alpha= 0.8)
img = mpimg.imread(os.path.join(os.getcwd(), 'plots', 'IsoCMD_comparison.png'))
plt.figure(figsize=(24,14))
plt.imshow(img)
plt.axis('off')
```

+ Non-rotating Geneva models are more red and less luminous than non-rotating MIST models (why?).

+ However, rotating Geneva models are more blue and more luminous than rotation MIST models, due to greater rotational mixing efficiency in Geneva models.

+ Geneva models go bluewards and more luminous as rotation increases; MIST models, up to a point go redwards and more luminous. At older ages, with lower masses on the MSTO, MIST models get redder and less luminous (shown below).

```python
rots = [0.0, 0.1, 0.3, 0.6]*2
num_isos=len(rots)
fehs = [0.0, 0.0, 0.0, 0.0]*2
ages = [8.2, 8.2, 8.2, 8.2, 8.8, 8.8, 8.8, 8.8]
pgc.compare_iso(fehs, rots, ages, [('log_Teff','log_L')], pltmass=False, geneva_on = False, ylim= [0.95, 3.0], xlim = [4.2, 3.85], alpha= 0.8)
img = mpimg.imread(os.path.join(os.getcwd(), 'plots', 'IsoCMD_comparison.png'))
plt.figure(figsize=(24,14))
plt.imshow(img)
plt.axis('off')
```

+ So due to non-rotating models being bluer than Geneva, we might expect higher metallicities than what BH15 found with Geneva in those cases.

+ Due to rotating models being redder, we might expect lower metallicities at higher rotation rates.

+ Rotation making models more red may also cause models to tend toward younger ages than BH15 at high rotation rates, and older ages at low rotation rate.

This appears to be what is seen in the MATCH best-fits.
