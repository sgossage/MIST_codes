```python
from MIST_codes.scripts import read_mist_models as rmm
import matplotlib.pyplot as plt

import numpy as np

%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize=(16,9))

# kwargs for plot legend and (anchored) text.
lgdkwargs = {"frameon":False,"prop":dict(size=18),"loc":3}
txtkwargs = {"frameon":False,"prop":dict(size=18),"loc":4}

age= 8.82

iso = rmm.ISOCMD(0.15, 0.4, ebv=0.003, exttag='mlta2.00')
iso.set_isodata(age, "Tycho_B-Tycho_V", "Tycho_V", 3.34, lgdlbl=r"$\alpha_{MLT}=2.00$")
#labels=['exttag'],texts=['age','feh','vvc','inc'])
iso.isoplot(ax, shade = 0.4, title=False, obs='hyades_debruTIGS_TychoBV')
iso = rmm.ISOCMD(0.15, 0.4, ebv=0.003, exttag='TP')
iso.set_isodata(age, "Tycho_B-Tycho_V", "Tycho_V", 3.34, lgdlbl=r"$\alpha_{MLT}=1.82$")
#labels=['exttag'])
iso.isoplot(ax, shade = 0.3, title=False)
iso = rmm.ISOCMD(0.15, 0.4, ebv=0.003, exttag='mlta1.60')
iso.set_isodata(age, "Tycho_B-Tycho_V", "Tycho_V", 3.34, lgdlbl=r"$\alpha_{MLT}=1.60$",texts=['age','feh','vvc','inc'])
#labels=['exttag'])
iso.isoplot(ax, shade = 0.2, label=txtkwargs, legend=lgdkwargs, title=False, ylims=[10,2], xlims=[0.0,2.0])

ax.set_aspect(np.abs((np.diff(ax.get_xlim()))/(np.diff(ax.get_ylim()))))

plt.savefig("hyades_alphamlt_vary.png", bbox_inches='tight', dpi=300)

plt.show()
```
