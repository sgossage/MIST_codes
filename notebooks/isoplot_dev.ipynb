```python
from MIST_codes.scripts import read_mist_models as rmm
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize=(12,8))

# kwargs for plot legend and (anchored) text.
lgdkwargs = {"frameon":False,"prop":dict(size=10),"loc":3}
txtkwargs = {"frameon":False,"prop":dict(size=10),"loc":4}
textlbl = "A\nB\nC"
age= 8.80

iso = rmm.ISOCMD(0.15, 0.0, ebv=0.001, exttag='TP')
iso.set_isodata(age, "2MASS_J-2MASS_Ks", "2MASS_Ks", 3.34, 
                lgdlbl=r"$\frac{\Omega}{\Omega_c}=0.6$",texts=textlbl)
_=iso.isoplot(ax, shade = None, label=txtkwargs, legend=lgdkwargs, title=False, ylims=[8,0], 
            c='k', ls=':')
```
