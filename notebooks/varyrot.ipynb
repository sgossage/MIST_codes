```python
from MIST_codes.scripts import read_mist_models as rmm
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(16,9))

# kwargs for plot legend and (anchored) text.
lgdkwargs = {"frameon":False,"prop":dict(size=14),"loc":2}
txtkwargs = {"frameon":False,"prop":dict(size=14),"loc":4}

oldage= 8.82
youngage=8.10

iso = rmm.ISOCMD(0.15, 0.0, ebv=0.003, exttag='TP')
iso.set_isodata(8.90, "Tycho_B-Tycho_V", "Tycho_V", 3.34, lgdlbl=r"$\frac{\Omega}{\Omega_c}=0.0,$"+" age = {:d} Myr".format(int((10**8.90)/1e6)))
x,y,dline00 = iso.isoplot(ax1, shade = None, title=False, ls='--', c='r')

iso = rmm.ISOCMD(0.15, 0.6, ebv=0.003, exttag='TP')
iso.set_isodata(oldage, "Tycho_B-Tycho_V", "Tycho_V", 3.34, lgdlbl=r"$\frac{\Omega}{\Omega_c}=0.6$")
x,y,sline06 = iso.isoplot(ax1, title = False, c = 'b')

iso = rmm.ISOCMD(0.15, 0.0, ebv=0.003, exttag='TP')
iso.set_isodata(oldage, "Tycho_B-Tycho_V", "Tycho_V", 3.34, lgdlbl=r"$\frac{\Omega}{\Omega_c}=0.0$",texts=['age','feh','inc'])
x,y,sline00 = iso.isoplot(ax1, label=txtkwargs, legend=lgdkwargs, title=False, ylims=[8,1.6],c = 'r')

ax1leg1 = ax1.legend(handles=[dline00], prop={'size':14}, loc=2)
ax1.add_artist(ax1leg1)
ax1.legend(handles=[sline00, sline06], prop={'size':14}, bbox_to_anchor=(0.95, 0.35))

#---------------------------------------

iso = rmm.ISOCMD(0.15, 0.0, ebv=0.003, exttag='TP')
iso.set_isodata(8.00, "Tycho_B-Tycho_V", "Tycho_V", 3.34, lgdlbl=r"$\frac{\Omega}{\Omega_c}=0.0,$"+" age = {:d} Myr".format(int((10**8.00)/1e6)))
x,y,dline00 = iso.isoplot(ax2, shade = None, title=False, ls='--', c='r')

iso = rmm.ISOCMD(0.15, 0.6, ebv=0.003, exttag='TP')
iso.set_isodata(youngage, "Tycho_B-Tycho_V", "Tycho_V", 3.34, lgdlbl=r"$\frac{\Omega}{\Omega_c}=0.6$")
x,y,sline06 = iso.isoplot(ax2, title=False, c='b')

iso = rmm.ISOCMD(0.15, 0.0, ebv=0.003, exttag='TP')
iso.set_isodata(youngage, "Tycho_B-Tycho_V", "Tycho_V", 3.34, lgdlbl=r"$\frac{\Omega}{\Omega_c}=0.0$",texts=['age','feh','inc'])
x,y,sline00 = iso.isoplot(ax2, label=txtkwargs, legend=lgdkwargs, title=False, ylims=[5,-1.2], c='r')

ax2leg1 = ax2.legend(handles=[dline00], prop={'size':14}, loc=2)
ax2.add_artist(ax2leg1)
ax2.legend(handles=[sline00, sline06], prop={'size':14}, bbox_to_anchor=(0.95, 0.35))

ax1.set_aspect(np.abs((np.diff(ax1.get_xlim()))/(np.diff(ax1.get_ylim()))))
ax2.set_aspect(np.abs((np.diff(ax2.get_xlim()))/(np.diff(ax2.get_ylim()))))

plt.savefig("vvc_vary.png", bbox_inches='tight', dpi=300)

plt.show()
```
