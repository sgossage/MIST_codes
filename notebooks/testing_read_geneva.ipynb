```python
from MIST_codes.scripts import read_geneva as rg
from MIST_codes.scripts import read_mist_models as rmm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iso = rg.iso(lage=8.80000)
iso2 = rg.iso(lage=8.80000,vvc=0.568)
iso3 = rg.iso(lage=8.70000)
miso = rmm.ISOCMD(0.00, 0.6, ebv=0.000, exttag='TP')
miso.set_isodata(8.8, "log_Teff", "log_L", 0.00, lgdlbl="MIST/MESA "+r"$\frac{\Omega}{\Omega_c}=0.6$"+', {:d} Myr'.format(int((10**8.8)/(1e6))) )

miso0 = rmm.ISOCMD(0.00, 0.0, ebv=0.000, exttag='TP')
miso0.set_isodata(8.8, "log_Teff", "log_L", 0.00, lgdlbl="MIST/MESA "+r"$\frac{\Omega}{\Omega_c}=0.0$"+', {:d} Myr'.format(int((10**8.8)/(1e6))) )

miso0_old = rmm.ISOCMD(0.00, 0.0, ebv=0.000, exttag='TP')
miso0_old.set_isodata(8.85, "log_Teff", "log_L", 0.00, lgdlbl="MIST/MESA "+r"$\frac{\Omega}{\Omega_c}=0.0$"+', {:d} Myr'.format(int((10**8.85)/(1e6))) )
```

```python
f, ax = plt.subplots(figsize=(12,8))

lgdkwargs = {"frameon":False,"prop":dict(size=14),"loc":4}
txtkwargs = {"frameon":False,"prop":dict(size=14),"loc":1}

iso.plot('logTe_c', 'logL', ax=ax, ls='-', c='b', label='Geneva '+r'$\frac{\Omega}{\Omega_c}=0.568$'+', {:d} Myr'.format(int((10**8.8)/(1e6))) )
iso2.plot('logTe_c', 'logL', ax=ax, ls=':',c='b', label='Geneva '+r'$\frac{\Omega}{\Omega_c}=0.0$')
miso.isoplot(ax, shade = None, title=False, c='r', ls=':',label=txtkwargs, legend=lgdkwargs)
miso0.isoplot(ax, shade = None, title=False, c='r', ls='-',label=txtkwargs, legend=lgdkwargs)
#plt.plot(iso.data['B-V'], iso.data['MV'])
#plt.plot(iso2.data['B-V'], iso2.data['MV'])
ax.set_ylim([0.7,4])
ax.set_xlim([4.2,3.4])
plt.savefig('GenMISTHRD_vvc0.0vs0.6.png', dpi=300)
#ax.invert_xaxis()
plt.show()
```

```python
f, ax = plt.subplots(figsize=(12,8))
iso.plot('logTe_c', 'logL', ax=ax, ls='-', c='r', label='Geneva '+r'$\frac{\Omega}{\Omega_c}=0.0$'+', {:d} Myr'.format(int((10**8.8)/(1e6))) )
iso2.plot('logTe_c', 'logL', ax=ax, ls='--',c='b', label='Geneva '+r'$\frac{\Omega}{\Omega_c}=0.568$'+', {:d} Myr'.format(int((10**8.8)/(1e6))) )
iso3.plot('logTe_c', 'logL', ax=ax, ls='-',c='k', label='Geneva '+r'$\frac{\Omega}{\Omega_c}=0.0$'+', {:d} Myr'.format(int((10**8.7)/(1e6))) )
#plt.plot(iso.data['B-V'], iso.data['MV'])
#plt.plot(iso2.data['B-V'], iso2.data['MV'])
ax.set_ylim([0,3])
ax.set_xlim([4.0,3.6])
ax.set_xlabel(r'$logT_{eff}$')
ax.set_ylabel(r'$logL/L_{\odot}$')
ax.legend(**lgdkwargs)
plt.savefig('GenHRD_vvc0.0vs0.6.png', dpi=300)
#ax.invert_xaxis()

plt.show()
```
