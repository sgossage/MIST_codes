```python
from MIST_codes.scripts import read_mist_models as rmm
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize=(12,8))

# kwargs for plot legend and (anchored) text.
lgdkwargs = {"frameon":False,"prop":dict(size=10),"loc":3}
txtkwargs = {"frameon":False,"prop":dict(size=10),"loc":4}

age= 8.83

iso = rmm.ISOCMD(0.15, 0.0, ebv=0.001, exttag='TP')
iso.set_isodata(age, "Bessell_B-Bessell_V", "Bessell_V", 3.34, lgdlbl=r"$\frac{\Omega}{\Omega_c}=0.0$")
#labels=['exttag'],texts=['age','feh','vvc','inc'])
iso.isoplot(ax, shade = None, title=False, obs='hyades_deB01_BV', c='k')
iso = rmm.ISOCMD(0.15, 0.4, ebv=0.001, exttag='TP')
iso.set_isodata(age, "Bessell_B-Bessell_V", "Bessell_V", 3.34, lgdlbl=r"$\frac{\Omega}{\Omega_c}=0.4$")
#labels=['exttag'])
iso.isoplot(ax, shade = None, title=False, c='k', ls=':')
iso = rmm.ISOCMD(0.15, 0.6, ebv=0.001, exttag='TP')
iso.set_isodata(age, "Bessell_B-Bessell_V", "Bessell_V", 3.34, lgdlbl=r"$\frac{\Omega}{\Omega_c}=0.6$",texts=['age','feh','inc'])
#labels=['exttag'])
iso.isoplot(ax, shade = None, label=txtkwargs, legend=lgdkwargs, title=False, ylims=[10,2], c='k', ls=':')

plt.show()
```
