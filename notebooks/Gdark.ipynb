```python
from MIST_codes.scripts import read_mist_models as rmm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize=(12,8))

# kwargs for plot legend and (anchored) text.
lgdkwargs = {"frameon":False,"prop":dict(size=10),"loc":3}
txtkwargs = {"frameon":False,"prop":dict(size=10),"loc":4}

age= 8.82

iso = rmm.ISOCMD(0.00, 0.4, ebv=0.0, exttag='TP', gravdark_i = 0)
iso.set_isodata(age, "Bessell_B-Bessell_V", "Bessell_V", 3.34, lgdlbl=r"$i=0^{\circ}$")
_ = iso.isoplot(ax, shade = 0.1, ylims=[10,2])

iso = rmm.ISOCMD(0.00, 0.4, ebv=0.0, exttag='TP', gravdark_i = 90)
iso.set_isodata(age, "Bessell_B-Bessell_V", "Bessell_V", 3.34, lgdlbl=r"$i=90^{\circ}$",texts=['age','feh','vvc'])
_= iso.isoplot(ax, shade = 0.2, label=txtkwargs, legend=lgdkwargs, title=False, ylims=[10,2])
```
