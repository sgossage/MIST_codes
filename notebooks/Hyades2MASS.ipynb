```python
from MIST_codes.scripts import read_mist_models as rmm
import matplotlib.pyplot as plt
import matplotlib as mpl
params = {'axes.labelsize': 20,'axes.titlesize':20, 'font.size': 14, 'legend.fontsize': 14, 'xtick.labelsize': 14, 'ytick.labelsize': 14}
mpl.rcParams.update(params)

%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize=(12,8))

# kwargs for plot legend and (anchored) text.
lgdkwargs = {"frameon":False,"prop":dict(size=12),"loc":2}
txtkwargs = {"frameon":False,"prop":dict(size=10),"loc":4}

age= 8.98

iso = rmm.ISOCMD(0.15, 0.3, ebv=0.001, exttag='TP')
iso.set_isodata(8.70, "2MASS_J-2MASS_Ks", "2MASS_Ks", 3.34, lgdlbl=r"$\frac{\Omega}{\Omega_c}=0.3$, [Fe/H]=0.15, age=631 Myr")
iso.isoplot(ax, shade = None, title=False, c='r', ls='--')

iso = rmm.ISOCMD(0.05, 0.0, ebv=0.001, exttag='TP')
iso.set_isodata(age, "2MASS_J-2MASS_Ks", "2MASS_Ks", 3.34, lgdlbl=r"$\frac{\Omega}{\Omega_c}=0.0$, [Fe/H]=0.05, age=940 Myr",texts=['age','feh','inc'])
#labels=['exttag'])
iso.isoplot(ax, shade = None, label=txtkwargs, legend=lgdkwargs, title=False, xlims=[-0.1, 0.8], ylims=[6,0], c='k', ls='-', obs='hyades_goldman13HIFI_2MASSJK')

plt.show()
```
